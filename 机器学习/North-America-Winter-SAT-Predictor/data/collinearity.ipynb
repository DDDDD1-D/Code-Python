```python
import sys
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

exclude = ['PC1','GlobeSST']

df = pd.read_csv("../data/PC1-input.csv")

features = [feature for feature in list(df.keys()) if feature not in exclude]
data = df[features]

corr = data.corr()

fig = plt.figure(figsize=(10, 10))

sns.set()

ax = sns.heatmap(corr, annot=True,vmax=1, vmin=-1, robust=True, cmap='RdBu')


plt.title("CORR")
plt.show()
```
