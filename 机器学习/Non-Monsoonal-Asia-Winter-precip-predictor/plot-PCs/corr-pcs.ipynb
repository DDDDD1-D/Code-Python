```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
PC = "PC1"
df = pd.read_csv("%s.csv" % PC)

corr = df.corr()

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

plt.subplots(figsize=(9, 9))
mymap = sns.heatmap(np.round(corr,2), annot=True, vmax=1, vmin=-1, square=True, cmap="RdBu_r", \
            linewidths=.5, mask=mask, annot_kws={'fontsize':'xx-large'}, cbar=False)

plt.tick_params(labelsize=14)
plt.title("(a) %s" % PC,fontsize=28)

cb = mymap.figure.colorbar(mymap.collections[0])
cb.ax.tick_params(labelsize=16)

plt.savefig("%s.png" % PC,dpi=1000)
```
