```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("PC1-importance.csv")
name = [feature for feature in list(df.keys())]

df
```

```python
plt.pie(x=df.iloc[0][1:9],labels=name[1:9],autopct='%.2f%%',pctdistance=0.8)
plt.savefig('PC1-XGBoost.png', dpi=1000)
```

```python
keys = [x for x in name if x != "SC-NA" and x != "nina4" and x != "method"]
plt.pie(x=df.iloc[1][keys],labels=keys,autopct='%.2f%%',pctdistance=0.8)
plt.savefig('PC1-LightGBM.png', dpi=1000)
```
